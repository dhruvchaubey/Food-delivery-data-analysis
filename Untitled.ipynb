```python
import pandas as pd
#--- Read in dataset ----
df =pd.read_csv('zomato.csv')

#--- Inspect data ---
df
```

**Dropping Unneccessary Columns**

For our analysis, we dont need address and phone number column. So your task it drop these two columns from our dataframe.

```python

df.drop(['address','phone'], axis=1, inplace = True)

#--- Inspect data ---
df
```

**Renaming Columns for Clarity**

Our task is to rename certain columns to improve readability. Specifically, you'll change rate to rating, approx_cost(for two people) to approx_cost, and listed_in(type) to type. This will make the column names more intuitive for further analysis.

```python
df.rename(columns={'rate': 'rating', 'approx_cost(for two people)': 'approx_cost', 'listed_in(type)': 'type'},
                  inplace=True)
#inspect the data
df
```

**Dropping Rows with Missing Values**

In this task, we'll remove any rows from the dataset where the 'name' column has missing values. This ensures that our analysis only includes entries with valid and complete data in the 'name' field.

```python
df.dropna(subset=['name'], inplace= True)

#--- Inspect data ---
df
```

**Handling Missing Values**

This section of code addresses missing values in the dataset by filling them with appropriate replacements based on the nature of each column.

**Key Points:**

**fillna() Function:** This method is used to replace missing values (NaN) in a DataFrame or Series.

**inplace=True:** Updates the original DataFrame directly, eliminating the need to reassign the result.

**Data Integrity:** Filling missing values ensures the dataset is consistent and ready for further analysis without interruptions or errors.

By systematically handling missing values, this code ensures the dataset is prepared for robust analysis and reduces the risk of inaccuracies in subsequent computations.


```python
# Dictionary to map columns to their respective fill values
fill_values = {
    'book_table': 'NA',
    'location': 'NA',
    'dish_liked': 'NA',
    'cuisines': 'NA',
    'rating': 0,
    'votes': 0,
    'approx_cost': 0,
    'rest_type': 'NA',
    'type': 'NA'
}

# Fill missing values in one step
df.fillna(value=fill_values, inplace=True)

#inspect data
df
```

**Purifying Zomato's Culinary Canvas: Removing Duplicates**

Our exploration of the Zomato dataset continues as we embark on a mission to eliminate duplicates. By removing redundancy, we refine our data, ensuring that each restaurant shines uniquely, guiding us toward the essence of exceptional dining.

**Purpose:** This line removes duplicate rows from the DataFrame, ensuring that each row is unique.

**Parameters:**

inplace=True: Modifies the original DataFrame directly without creating a new one.

keep='first': Retains the first occurrence of each duplicated row and removes subsequent duplicates.

**Why:** Ensures data integrity by eliminating redundancy, which is crucial for accurate analysis and modeling.

```python
df.drop_duplicates(inplace=True, keep= 'first')

#--- Inspect data ---
df
```

**Data Filtering**

This section of code is designed to filter out rows from the DataFrame (df) where specific columns contain unwanted strings, specifically 'RATED' or 'Rated'.

**Objective:**

To clean the dataset by removing rows with irrelevant or inconsistent entries in the specified columns: 'name', 'type', 'approx_cost','cuisines','dish_liked','rest_type','location','votes','rating','book_table'. This step ensures that the data is accurate, consistent, and suitable for further analysis.




```python
# Function to filter out rows containing 'RATED' or 'Rated' in specified columns
filter_columns = ['name', 'type', 'approx_cost','cuisines','dish_liked','rest_type','location','votes','rating','book_table']
for col in filter_columns:
    df = df[~df[col].str.contains('RATED|Rated', na=False)]
    
#inspect the data
df
```

**Filtering Rows with 'Yes' or 'No' in the 'Online Order' Column**

Our task is to keep only the rows where the 'online_order' column contains 'Yes' or 'No'. This step filters the dataset to include only entries with valid online ordering information, ensuring that the analysis is based on relevant and accurate data.

```python
df = df[df['online_order'].str.contains('Yes|No') == True]

#--- Inspect data ---
df
```

**Replacing Values and Cleaning the 'Rating' Column**

What Are We Doing?
In this step, we are cleaning and standardizing the values in the rating column of the dataset by performing the following actions:

**Replacing Placeholder Values:**

Replace special strings like 'NEW' and '-' with 0.
These placeholder values typically indicate missing or unavailable ratings. Converting them to 0 ensures numerical consistency in the column.

Removing the /5 Suffix:
Many ratings might be represented as strings like '4.5/5'. Removing the /5 suffix ensures that only the numerical rating value (e.g., 4.5) remains.

This step transforms the rating column into a format suitable for numerical analysis.

```python
# Make a copy of the DataFrame or column to avoid the SettingWithCopyWarning
df = df.copy()

# Replace 'NEW' and '-' with 0, and remove '/5' suffix from the 'rating' column
df['rating'].replace({'NEW': 0, '-': 0, '/5': ''}, regex=True, inplace=True)

# Inspect the data
df

```
